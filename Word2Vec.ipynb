```python
import gensim
```

```python
model = gensim.models.Word2Vec.load('./ko/ko.bin')
```

```python
query = ['한국','서울','파리']
x = model.wv.similar_by_vector(model.wv['한국'] - model.wv['서울'] + model.wv['파리'])
x = [k for k in x if k[0] not in query][:3]
x
```
